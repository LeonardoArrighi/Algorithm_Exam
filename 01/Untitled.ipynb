```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

a = []

for i in range(1,13):
    a.append(2**i)
    
naive = [0.000001, 0.000001, 0.000002, 0.000008, 0.000066, 0.000313, 0.003285, 0.024794, 0.194574, 2.245232, 52.637351, 444.416258]

str = [0.000001, 0.000000, 0.000001, 0.000007, 0.0000046, 0.000406, 0.002943, 0.017182, 0.143421, 0.754815, 5.581592, 36.970885]

str_opt = [0.000001, 0.000000, 0.000001, 0.000007, 0.0000048, 0.000396, 0.002653, 0.015847, 0.120413, 0.708622, 5.222716, 35.056560]

print(a)

fig, ax = plt.subplots(1, 1, figsize=(15, 18))

ax.plot(a, naive, color='b', label = 'Naive algorithm')
ax.plot(a, naive, color='b', marker='o')

ax.plot(a, str, color='orange', label = 'Strassen algorithm with peeling')
ax.plot(a, str, color='orange', marker='o')

ax.plot(a, str_opt, color='r', label = 'Optimized Strassen algorithm with peeling')
ax.plot(a, str_opt, color='r', marker='o')

ax.grid()
ax.legend()
ax.set_title('Time vs. Size of matrix')
```
